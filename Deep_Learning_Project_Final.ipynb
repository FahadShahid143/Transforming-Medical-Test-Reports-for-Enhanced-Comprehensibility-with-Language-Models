```python
import pandas as pd
import numpy as np
import re
import pdfplumber
import openai
import gradio as gr
```

```python
## FUNCTION FOR PDF PARSER - EXTRACTING INFORMATION FROM PDF FILE
```

```python
def extract_pdf_data(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        page = pdf.pages[0]
        text = page.extract_text()

        # Initialize data collection
        data = []
        report_name = None

        # Try to identify report name by analyzing the format and content of lines
        lines = text.split('\n')
        for i, line in enumerate(lines):

            if "Report" in line or "Count" in line or "Summary" in line:
                if len(line.split()) < 10:
                    report_name = line
                    break

        # Regex to capture the data table entries
        pattern = re.compile(r"(\D+?)\s+([\d.]+)\s*(Low|High|Borderline)?\s*([\d\s.-]+)?\s*(%|g/dL|mill/cumm|fl|pg|cumm)")

        # Extracting data from structured part
        for line in lines:
            match = pattern.search(line)
            if match:
                data.append({
                    'Investigation': match.group(1).strip(),
                    'Result': match.group(2).strip(),
                    'Reference Value': (match.group(3) if match.group(3) else '') + ' ' + (match.group(4) if match.group(4) else '').strip(),
                    'Unit': match.group(5).strip()
                })

        # Create DataFrame
        df = pd.DataFrame(data)
        return df, report_name
```

```python
import pytesseract
from PIL import Image
```

```python
# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'
```

```python
def extract_image_data(image_path):
  image = Image.open(image_path)

  # Perform OCR on the image
  extracted_text = pytesseract.image_to_string(image)

  # Define the regex pattern
  pattern = re.compile(r"(\D+?)\s+([\d.]+)\s*(Low|High|Borderline)?\s*([\d\s.-]+)?\s*(%|g/dL|mill/cumm|fL|pg|cumm)")

  matches = pattern.findall(extracted_text)

  investigations = []
  results = []
  reference_values = []
  units = []

  for match in matches:
      investigations.append(match[0].strip())
      results.append(match[1])
      reference_values.append(match[3].strip() if match[3] else '')
      units.append(match[4])

  # Creating dataframe
  data = {
      'Investigation': investigations,
      'Result': results,
      'Reference Value': reference_values,
      'Unit': units
  }

  df = pd.DataFrame(data)

  # Return the extracted information
  return df
```

```python
def Inference(prompt):
    promptGptQuestionClassifier = f"""
    Using the following lab test results generate a summary of the test report in the following format:
    Here is the lab test report:
        {prompt}

        Your response should be in the following format:
        Summary:
        (A small summary of the current situation)

        Deviations:
        (Which values are deviating from the normal values)

        Criticality:
         (What is the crticality of the situation)

        """
    response = client.completions.create(
      model="gpt-3.5-turbo-instruct",
      prompt = promptGptQuestionClassifier,
      max_tokens = 400,
      temperature = 0.15,
      )
    generatedAnswer = response.choices[0].text
    return generatedAnswer
```
